```python
from utils import get_path,read_jsonl
import matplotlib.pyplot as plt
from PIL import Image as PIL_Image
def show_img(data_dir,index):
    data=read_jsonl(data_dir)
    item=data[index-1]
    id=item['image_id']
    image_path=get_path(id)
    img = PIL_Image.open(image_path)
    plt.axis('off')
    plt.imshow(img)
    # plt.imshow(image_path)
```
