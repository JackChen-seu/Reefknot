```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Data
labels = ['Poster', 'Color', 'Position', 'Count', 'Existence', 'Code Reasoning',
          'Text Translation', 'Numerical Calculation', 'OCR', 'Commonsense Reasoning',
          'Artwork', 'Landmark', 'Scene', 'Celebrity']
categories = len(labels)

# Define the data for each model
WeMM = [75, 80, 90, 85, 100, 95, 80, 60, 50, 70, 65, 80, 75, 85]
Lion = [65, 75, 85, 80, 95, 90, 70, 55, 45, 65, 60, 75, 70, 80]
SPHINX = [70, 78, 88, 83, 98, 93, 75, 58, 48, 68, 63, 78, 73, 83]
GPT_4V = [80, 85, 95, 90, 105, 100, 85, 65, 55, 75, 70, 85, 80, 90]
InfMLLM = [60, 70, 80, 75, 90, 85, 65, 50, 40, 60, 55, 70, 65, 75]
MMICL = [55, 65, 75, 70, 85, 80, 60, 45, 35, 55, 50, 65, 60, 70]

# Combine all data
data = [WeMM, Lion, SPHINX, GPT_4V, InfMLLM, MMICL]

# Number of models
num_models = len(data)

# Create radar chart
angles = np.linspace(0, 2 * np.pi, categories, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

for i in range(num_models):
    values = data[i] + data[i][:1]
    ax.fill(angles, values, alpha=0.25)
    ax.plot(angles, values, label=f'Model {i+1}')

# Add labels
plt.xticks(angles[:-1], labels)

# Add legend
ax.legend(['WeMM', 'Lion', 'SPHINX', 'GPT-4V', 'InfMLLM', 'MMICL'], loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Show the plot
plt.show()
```
